# Descriptive Statistics Using Pandas of Kaggle Mental Health Dataset

In this notebook, we will explore a dataset about student mental health and will evaluate how different groups of students report their mental health status.

## Load dataset

```python
import lib
import pandas as pd
```

```python
# get path from current directory to data folder
data = pd.read_csv("./data/student_mental_health.csv")
data.head()
```

## Descriptive Statistics

We look at overall descriptive statistics from the dataset. These include:
- Proportion of male vs. female respondants
- Age: Mean, Median, and Mode
- Course of study distribution
- Year of study distribution
- Marital status distribution
- Anxiety panic attack, and specialist help distribution

#### Proportion of male vs. female respondants

```python
lib.bar_chart(
    data,
    "Choose your gender",
    "Gender Distribution",
    "Gender",
    "Count",
    color=["pink", "skyblue"],
)
```

#### Age: Mean, Median, Mode

```python
# from lib import get_mean_median_mode
# Get mean, median, mode from lib module
mean_val, mode_val, median_val = lib.get_mean_median_mode(data, "Age")

# print results
print("Mean age: ", mean_val)
print("Median age: ", median_val)
print("Mode age: ", mode_val)
```

#### Age: 25th, 50th, 75th percentiles

```python
# get percentiles
percentile_25, percentile_50, percentile_75 = lib.get_percentiles(data, "Age")

# print results
print("25th percentile: ", percentile_25)
print("50th percentile: ", percentile_50)
print("75th percentile: ", percentile_75)
```
